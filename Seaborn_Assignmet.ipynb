```python
# Run this code cell
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

url = 'https://new-assets.ccbp.in/frontend/content/aiml/classical-ml/co2_emissions_canada.csv'

# Upload the data on colab and execute this code cell
data = pd.read_csv(url)
data.head()
```

**1. Find the top 3 vehicles with the highest CO2 emissions for each vehicle class.**

```python
top_3_co2_emissions = data.groupby('Vehicle Class').apply(lambda x: x.nlargest(3, 'CO2 Emissions(g/km)')).reset_index(drop=True)
display(top_3_co2_emissions)
```

**2. Find the top 3 car makes with the highest average CO2 emissions.**

```python
# Write the code here

top_3_makes = data.groupby('Make').apply(lambda x: x.nlargest(3, 'CO2 Emissions(g/km)')).reset_index(drop=True)
display(top_3_makes)
```

**3. Identify the fuel type that has the lowest average CO2 emissions.** In the given format:
- `Fuel type with lowest average CO2 emissions: N`

```python
# Write the code here

lowest_emission_fuel = data.groupby('Fuel Type').agg({'CO2 Emissions(g/km)': 'mean'}).idxmin()
print(f"Fuel type with lowest average CO2 emissions: {lowest_emission_fuel.iloc[0]}")
```

**4. Add a new column for CO2 emissions per liter of fuel consumed (CO2 Emissions / Fuel Consumption) and list first 10 rows**

```python
# Write the code here

data['CO2 Emissions per liter of fuel consumed'] = data['CO2 Emissions(g/km)'] / data['Fuel Consumption Comb (L/100 km)']
data.head(10)
```

**5. Find the car make with the highest fuel consumption on average.**

```python
# Write the code here

highest_fuel_make = data.groupby('Make')['Fuel Consumption Comb (L/100 km)'].mean().idxmax()
print(f"Car make with the highest average fuel consumption: {highest_fuel_make}")
```

**6. Identify the top 3 car makes that have the most models with CO2 emissions above the 75th percentile.**

```python
# Write the code here

threshold = data['CO2 Emissions(g/km)'].quantile(0.75)
top_makes = data[data['CO2 Emissions(g/km)'] > threshold]['Make'].value_counts().nlargest(3)
print("Top 3 car makes with most models above 75th percentile CO2 emissions:")
print(top_makes)
```

**7. Create a new feature that categorizes vehicles into 'Low', 'Medium', and 'High' based on percentiles of the "CO2 Emissions(g/km)" column and get first 10 rows:**

- `Low: ≤ 33rd percentile`
- `Medium: 33rd to 66th percentile`
- `High: > 66th percentile`

```python
# Write the code here

lower_bound = data['CO2 Emissions(g/km)'].quantile(0.33)
upper_bound = data['CO2 Emissions(g/km)'].quantile(0.66)

def categorize_co2(emission):
    if emission <= lower_bound:
        return 'Low'
    elif emission > lower_bound and emission <= upper_bound:
        return 'Medium'
    else:
        return 'High'

data['CO2 Emission Category'] = data['CO2 Emissions(g/km)'].apply(categorize_co2)
data.head(10)
```

**8. Determine if turbocharged engines (assuming turbocharged cars contain 'T' in model name) have higher average CO2 emissions.**

```python
# Write the code here

turbocharged_cars = data[data['Model'].str.contains('T', na=False)]
non_turbocharged_cars = data[~data['Model'].str.contains('T', na=False)]

avg_co2_turbo = turbocharged_cars['CO2 Emissions(g/km)'].mean()
avg_co2_non_turbo = non_turbocharged_cars['CO2 Emissions(g/km)'].mean()

print(f"Average CO2 emissions for turbocharged cars: {avg_co2_turbo:.2f} g/km")
print(f"Average CO2 emissions for non-turbocharged cars: {avg_co2_non_turbo:.2f} g/km")

if avg_co2_turbo > avg_co2_non_turbo:
    print("Turbocharged engines tend to have higher average CO2 emissions.")
else:
    print("Turbocharged engines do not tend to have higher average CO2 emissions.")
```

 **9. Compute the mean and standard deviation of CO2 emissions for each fuel type, and sort by highest variation.**

```python
# Write the code here

fuel_stats = data.groupby('Fuel Type')['CO2 Emissions(g/km)'].agg(['mean', 'std'])
sorted_fuel_stats = fuel_stats.sort_values(by='std', ascending=False)
display(sorted_fuel_stats)
```

**10. Categorize vehicles into engine size (5 bins), calculates the average CO2 emissions for each engine size bin and prints the engine size range that has the highest average CO2 emissions.**

```python
# Write the code here

data['Engine Size Bin'] = pd.cut(data['Engine Size(L)'], bins=5)
engine_emission = data.groupby('Engine Size Bin')['CO2 Emissions(g/km)'].mean()
highest_emission_bin = engine_emission.idxmax()
print(f"Engine size range with the highest average CO2 emissions: {highest_emission_bin}")
```

**11. Compare the median fuel consumption between vehicles with engine sizes above and below the median engine size.** In the given format:
- `Above median engine size fuel consumption: ____`
- `Below median engine size fuel consumption: ____`

```python
# Write the code here
median_engine_size = data['Engine Size(L)'].median()
above_median_fuel = data[data['Engine Size(L)'] > median_engine_size]['Fuel Consumption Comb (L/100 km)'].median()
below_median_fuel = data[data['Engine Size(L)'] <= median_engine_size]['Fuel Consumption Comb (L/100 km)'].median()

print(f"Above median engine size fuel consumption: {above_median_fuel}")
print(f"Below median engine size fuel consumption: {below_median_fuel}")
```

**12. Find the vehicle with the best CO2 emission per horsepower ratio (assuming 'Engines Size' correlates with power).**

```python
# Write the code here
data['CO2 per Engine Size'] = data['CO2 Emissions(g/km)'] / data['Engine Size(L)']
best_efficiency = data.loc[data['CO2 per Engine Size'].idxmin()]
display(best_efficiency)
```

**13. Find the average CO2 emissions for each make, considering only models with fuel consumption above 10 L/100 km.**

```python
high_fuel_cars = data[data['Fuel Consumption Comb (L/100 km)'] > 10]
avg_co2_per_make = high_fuel_cars.groupby('Make')['CO2 Emissions(g/km)'].mean().sort_values(ascending=False)
display(avg_co2_per_make)
```
